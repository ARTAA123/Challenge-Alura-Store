### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ventas_tienda1 = sum(tienda["Precio"])
ventas_tienda2 = sum(tienda2["Precio"])
ventas_tienda3 = sum(tienda3["Precio"])
ventas_tienda4 = sum(tienda4["Precio"])

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingreso_total_tienda = [ventas_tienda1, ventas_tienda2, ventas_tienda3, ventas_tienda4]
ingreso_total_tienda
```

```python
#Gráfico con matplotlib
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')
plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas, ingreso_total_tienda, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], edgecolor='black', linewidth=1.5)

# Agregar etiquetas de valores encima de las barras
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 500, f"${bar.get_height():,}",
             ha='center', va='bottom', fontsize=12, color='black')


# Personalización el gráfico
plt.title('Comparación de Ingresos totales por Tiendas', fontsize =16, weight='bold')
plt.xlabel('Tiendas', fontsize=14)
plt.ylabel('Ingresos (en USD)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

#Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría
En este espacio se realiza el calculo de la cantidad de ventas por categoria de producto

```python
categorias_productos = list(set(tienda['Categoría del Producto']))
# Categorias por producto
categorias_populares_tienda1 = {}
categorias_populares_tienda2 = {}
categorias_populares_tienda3 = {}
categorias_populares_tienda4 = {}

#Definición de función que agrega categoría a un diccionario y realiza el conteo de productos
def conteo_por_categoria(tienda, lista):
  for elemento in tienda['Categoría del Producto']:
    if elemento in lista:
      lista[elemento] += 1
    else:
      lista[elemento] = 1
```

```python
conteo_por_categoria(tienda, categorias_populares_tienda1)
conteo_por_categoria(tienda2, categorias_populares_tienda2)
conteo_por_categoria(tienda3, categorias_populares_tienda3)
conteo_por_categoria(tienda4, categorias_populares_tienda4)
```

# 3. Calificación promedio de la tienda

## Definición de funciones para obtener la calificación por categoria de producto y generar gráfico de barras por tienda consultada

```python
#Saca el promedio de calificación por tienda consultada
def promedio_categoria(data):
  promedios = data.groupby("Categoría del Producto")["Calificación"].mean().reset_index()
  promedios.columns =["Categoría del Producto", "Promedio de Calificación"]
  return promedios

#Genera una grafica de Barras para la tienda seleccionada
def grafica_promedios_producto(data):
  import matplotlib.pyplot as plt
  # Crear la gráfica
  plt.figure(figsize=(10, 6))  # Tamaño de la figura
  plt.bar(data['Categoría del Producto'], data["Promedio de Calificación"], color="skyblue", edgecolor="black")
  # Añadir etiquetas y título
  plt.title("Promedio de Calificaciones por Producto", fontsize=16)
  plt.xlabel("Categoría del Producto", fontsize=14)
  plt.ylabel("Promedio de Calificación", fontsize=14)
  plt.xticks(rotation=45, ha="right", fontsize=12)  # Rotar las etiquetas del eje X
  plt.grid(axis="y", linestyle="--", alpha=0.7)
  # Mostrar la gráfica
  plt.tight_layout()
  plt.show()
```

## Calificación de los usuarios por categoria de productos tienda 1

```python
promedio_categoria_tienda1 = promedio_categoria(tienda)
grafica_promedios_producto(promedio_categoria_tienda1)
```

# 4. Productos más y menos vendidos
- Se evaluan los productos con mas y menos ventas en cada tienda
- Se realiza la gráfica por tienda

```python
#Saca el promedio de ventas por producto en cada tienda
def promedio_venta_por_producto(data):
  total_venta = data.groupby("Producto")["Precio"].sum().reset_index()
  total_venta.columns =["Producto", "Ventas totales producto"]
  total_venta = total_venta.sort_values(by="Ventas totales producto", ascending=False)
  return total_venta

from matplotlib.ticker import FuncFormatter

# Función para formatear los números en el eje Y
def millones_formatter(x, pos):
  if x < 1:
    return f'{x:.2f}M'
  return f'{int(x)}M'

#Genera un gráfico con los 10 productos más vendidos
def producto_mas_vendidos(data, nombre):
  # Seleccionar los 10 productos más vendidos
  top_10_productos = data.head(10)

  # Gráfico de barras
  plt.figure(figsize=(10, 6))
  plt.bar(top_10_productos["Producto"], top_10_productos["Ventas totales producto"]/ 1_000_000, color="skyblue", edgecolor="black")

  # Añadir etiquetas y título
  plt.title(f'Top 10 Productos Más Vendidos en {nombre}', fontsize=16)
  plt.xlabel("Producto", fontsize=14)
  plt.ylabel("Cantidad Vendida(En Millones $)", fontsize=14)
  plt.xticks(rotation=45, ha="right", fontsize=12)  # Rotar etiquetas
  plt.gca().yaxis.set_major_formatter(FuncFormatter(millones_formatter))
  # Mostrar la gráfica
  plt.tight_layout()
  plt.show()

#Genera un gráfico con los 10 productos menos vendidos
def producto_menos_vendidos(data, nombre):
  # Seleccionar los 10 productos menos vendidos
  top_10_productos = data.tail(10)

  # Gráfico de barras
  plt.figure(figsize=(10, 6))
  plt.bar(top_10_productos["Producto"], top_10_productos["Ventas totales producto"]/ 1_000_000, color="skyblue", edgecolor="black")

  # Añadir etiquetas y título
  plt.title(f'Top 10 Productos Menos Vendidos en {nombre}', fontsize=16)
  plt.xlabel("Producto", fontsize=14)
  plt.ylabel("Cantidad Vendida (En millones $)", fontsize=14)
  plt.xticks(rotation=45, ha="right", fontsize=12)  # Rotar etiquetas

  plt.gca().yaxis.set_major_formatter(FuncFormatter(millones_formatter))

  # Mostrar la gráfica
  plt.tight_layout()
  plt.show()
```

```python
ventas_producto_tienda1 =promedio_venta_por_producto(tienda)
producto_mas_vendidos(ventas_producto_tienda1, 'Tienda 1')
```

```python
producto_menos_vendidos(ventas_producto_tienda1, 'Tienda 1')
```

```python
ventas_producto_tienda2 =promedio_venta_por_producto(tienda2)
producto_mas_vendidos(ventas_producto_tienda2, 'Tienda 2')
```

```python
producto_menos_vendidos(ventas_producto_tienda2, 'Tienda 2')
```

```python
ventas_producto_tienda3 =promedio_venta_por_producto(tienda3)
producto_mas_vendidos(ventas_producto_tienda3, 'Tienda 3')
```

```python
producto_menos_vendidos(ventas_producto_tienda3, 'Tienda 3')
```

```python
ventas_producto_tienda4 =promedio_venta_por_producto(tienda4)
producto_mas_vendidos(ventas_producto_tienda4, 'Tienda 4')
```
